```python
datos = pd.read_csv('conjunto_de_datos_iter_00CSV20.csv', encoding='utf-8')
datos.head()
```

```python
datos2 = datos[['NOM_MUN','NOM_LOC',	'PCON_LIMI','PHOG_IND','P5_HLI_NHE','PCON_DISC','VPH_PISODT','VPH_S_ELEC']]
```

```python
datos2
```

```python
datos2 = datos2[datos2['NOM_LOC'] == 'Total del Municipio']
```

```python
datos2
```

```python
# Asegurarse de que la columna 'NOM_MUN' sea de tipo cadena
datos2.loc[:, 'NOM_MUN'] = datos2['NOM_MUN'].astype(str)

# Crear una nueva columna para el nombre del municipio con sufijos numerados
datos2.loc[:, 'NOM_MUN_UNIQUE'] = datos2.groupby('NOM_MUN').cumcount() + 1

# Crear la nueva columna 'NOM_MUN_UNIQUE' con el formato deseado
datos2.loc[:, 'NOM_MUN_UNIQUE'] = datos2['NOM_MUN'] + datos2['NOM_MUN_UNIQUE'].astype(str)

# Ahora 'datos2' tendrá los nombres de los municipios con sufijos como 'Allende1', 'Allende2', etc.
datos2
```

```python
datos2
```

```python
# Asegurarse de que la columna 'municipio' sea de tipo cadena
vul.loc[:, 'municipio'] = vul['municipio'].astype(str)

# Crear una nueva columna para el nombre del municipio con sufijos numerados
vul.loc[:, 'municipio_count'] = vul.groupby('municipio').cumcount() + 1

# Crear una nueva columna de 'municipio' con el formato deseado
vul.loc[:, 'municipio_unique'] = vul['municipio'] + vul['municipio_count'].astype(str)

# Ahora 'vul' tendrá los nombres de los municipios con sufijos como 'Allende1', 'Allende2', etc.
vul
```

```python
datos2.columns
```

```python
df3 = pd.merge(vul, datos2, left_on='municipio_unique', right_on='NOM_MUN_UNIQUE', how='left')
```

```python
df3.drop(columns=['municipio_unique', 'NOM_MUN_UNIQUE','municipio_count'], inplace=True)
```
